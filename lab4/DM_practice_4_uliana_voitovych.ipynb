#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

**Об'єкти Нарендра-Партхізаратхі (Narendra-Parthasarathy)** — це нелінійні системи управління, які часто використовуються для тестування адаптивних і нелінійних регуляторів. Їх поведінка моделюється нелінійними диференціальними рівняннями, що описують динаміку в часі. Ці рівняння створюють цікаву, нелінійну та хаотичну поведінку системи.

### **Нелінійних рівнянь Нарендра-Партхізаратхі:**

1. $$
\dot{x}_1 = -a_1 x_1 + b_1 x_2^2
$$
$$
\dot{x}_2 = -a_2 x_2 + b_2 \sin(x_1)
$$

2. $$
\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2
$$
$$
\dot{x}_2 = a_2 x_1 x_2 - b_2 \cos(x_1)
$$

3. $$
\dot{x}_1 = -a_1 x_1 + b_1 x_2^3
$$
$$
\dot{x}_2 = a_2 \sin(x_1) - b_2 x_2
$$

4. $$
\dot{x}_1 = a_1 x_1 - b_1 x_2 \sin(x_1)
$$
$$
\dot{x}_2 = -a_2 x_2 + b_2 \cos(x_1)
$$

5. $$
\dot{x}_1 = -a_1 x_1 + b_1 x_2^2 \cos(x_1)
$$
$$
\dot{x}_2 = a_2 \sin(x_1) - b_2 x_1 x_2
$$

6. $$
\dot{x}_1 = a_1 x_1 x_2 - b_1 x_1^3
$$
$$
\dot{x}_2 = a_2 \sin(x_1) - b_2 x_2^2
$$

7. $$
\dot{x}_1 = -a_1 x_1^2 + b_1 \cos(x_2)
$$
$$
\dot{x}_2 = a_2 x_1 - b_2 \sin(x_2)
$$

8. $$
\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2
$$
$$
\dot{x}_2 = -a_2 x_2 + b_2 x_1 \cos(x_2)
$$

9. $$
\dot{x}_1 = a_1 x_1^2 - b_1 x_2^2
$$
$$
\dot{x}_2 = a_2 \cos(x_1) - b_2 \sin(x_2)
$$

10. $$
\dot{x}_1 = -a_1 x_1 \sin(x_2) + b_1 x_2^3
$$
$$
\dot{x}_2 = a_2 \cos(x_1) - b_2 x_1 x_2
$$

11. $$
\dot{x}_1 = a_1 \sin(x_1 x_2) - b_1 x_1
$$
$$
\dot{x}_2 = -a_2 x_2^2 + b_2 \cos(x_1)
$$

12. $$
\dot{x}_1 = -a_1 x_1 + b_1 \sin(x_2^2)
$$
$$
\dot{x}_2 = a_2 x_1^2 - b_2 \cos(x_1)
$$

13. $$
\dot{x}_1 = a_1 x_1 x_2 - b_1 \sin(x_1^2)
$$
$$
\dot{x}_2 = -a_2 x_2 + b_2 \cos(x_1 x_2)
$$

14. $$
\dot{x}_1 = a_1 x_1 - b_1 x_2^2 \sin(x_1)
$$
$$
\dot{x}_2 = a_2 \cos(x_1) - b_2 x_1^3
$$

15. $$
\dot{x}_1 = a_1 \cos(x_2) - b_1 x_1^2 \sin(x_1)
$$
$$
\dot{x}_2 = -a_2 x_1 x_2 + b_2 \cos(x_1^2)
$$

16. $$
\dot{x}_1 = -a_1 \sin(x_1) + b_1 x_2^2
$$
$$
\dot{x}_2 = a_2 \sin(x_1 x_2) - b_2 \cos(x_1)
$$

17. $$
\dot{x}_1 = a_1 x_1^2 \cos(x_2) - b_1 x_1 x_2
$$
$$
\dot{x}_2 = -a_2 x_2 + b_2 \sin(x_1^2)
$$

18. $$
\dot{x}_1 = -a_1 x_1 \sin(x_2) + b_1 \cos(x_1^2)
$$
$$
\dot{x}_2 = a_2 x_1^2 - b_2 \cos(x_1 x_2)
$$

19. $$

$$
\dot{x}_2 = a_2 \cos(x_1) - b_2 x_2^2
$$

20. $$
\dot{x}_1 = -a_1 x_1^2 + b_1 \sin(x_1 x_2)
$$
$$
\dot{x}_2 = a_2 \cos(x_1) - b_2 \sin(x_2^2)
$$

### Параметри:
- $( a_1, a_2 )$, $( b_1, b_2 )$ — це коефіцієнти, які можуть бути варійовані для дослідження динаміки системи та впливу на прогнозування.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor
```

```python
a1, a2, b1, b2 = 0.5, 0.8, 0.2, 0.6

dt = 0.01
num_points = 10000

x1 = np.zeros(num_points)
x2 = np.zeros(num_points)

x1[0], x2[0] = 1.0, 1.0

for i in range(1, num_points):
    dx1 = a1 * np.sin(x2[i-1]) - b1 * x1[i-1]**2
    dx2 = a2 * x1[i-1] * x2[i-1] - b2 * np.cos(x1[i-1])
    
    x1[i] = x1[i-1] + dx1 * dt
    x2[i] = x2[i-1] + dx2 * dt


np_data = pd.DataFrame({"x1": x1, "x2": x2})

np_data.to_csv("np_dataset.csv", index=False)
```

```python
data = pd.read_csv("np_dataset.csv")

train_data, temp_data = train_test_split(data, test_size=0.30, random_state=42)
val_data, test_data = train_test_split(temp_data, test_size=0.50, random_state=42)

scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data)
val_data_scaled = scaler.transform(val_data)
test_data_scaled = scaler.transform(test_data)
```

```python
lr_model = LinearRegression()
dt_model = DecisionTreeRegressor()
rf_model = RandomForestRegressor()
svm_model = SVR()
xgb_model = XGBRegressor()

X_train = train_data_scaled[:, :-1]
y_train = train_data_scaled[:, -1]

lr_model.fit(X_train, y_train)
dt_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
xgb_model.fit(X_train, y_train)
```

```python
dt_params = {
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5, 10]
}

dt_grid = GridSearchCV(dt_model, dt_params, cv=3)
dt_grid.fit(X_train, y_train)

rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5, 10]
}

rf_grid = GridSearchCV(rf_model, rf_params, cv=3)
rf_grid.fit(X_train, y_train)

svm_params = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

svm_grid = GridSearchCV(svm_model, svm_params, cv=3)
svm_grid.fit(X_train, y_train)

xgb_params = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 6, 10],
    'subsample': [0.6, 0.8, 1.0]
}

xgb_grid = GridSearchCV(xgb_model, xgb_params, cv=3)
xgb_grid.fit(X_train, y_train)
```

```python
X_val = val_data_scaled[:, :-1]
y_val = val_data_scaled[:, -1]
 
models = [lr_model, dt_grid.best_estimator_, rf_grid.best_estimator_, svm_grid.best_estimator_, xgb_grid.best_estimator_]
model_names = ["Linear Regression", "Decision Tree", "Random Forest", "SVM", "XGBoost"]

for model, name in zip(models, model_names):
    
    y_pred = model.predict(X_val)
    mse = mean_squared_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)
    print(f"{name} \nMSE: {mse}, \nR²: {r2}\n")
```

```python
X_test = test_data_scaled[:, :-1]
y_test = test_data_scaled[:, -1]

for model, name in zip(models, model_names):
    
    y_pred = model.predict(X_test)
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=y_test, y=y_pred)
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title(f"Actual vs Predicted - {name}")
    
    plt.show()
```

Було застосовано кілька моделей машинного навчання, зокрема лінійна регресія, дерево рішень, випадковий ліс, SVM і XGBoost. Моделі були навчені за допомогою навчального набору даних, при цьому 70% даних було зарезервовано для навчання, 15% для перевірки та 15% для тестування.

Оцінка моделей відбувалася за допомогою `MSE` (середня квадратична помилка) і `R²` як для перевірки, так і для тестових наборів даних, щоб виміряти прогнозовану ефективність і визначити найкращу модель.

1.  `Лінійна регресія` є простою та придатною для інтерпретації, але їй було нелегко охопити складність набору даних, що відображено у відносно низькій оцінці `R² (0,439)` і високому `MSE (0,554)`. Дана регресія припускає лінійний зв’язок між функціями та метою, що свідчить про її меншу ефективність для нелінійних шаблонів у даних.

2.  `Дерево рішень` перевершило лінійну регресію, досягнувши набагато вищого `R² (0,988)` і нижчого значення `MSE (0,0117)`. Дерева рішень як нелінійні моделі розбивають дані на менші регіони та підлаштовуються під модель у кожному з них. Незважаючи на потужність, вони можуть легко перенавчитися без належної регуляризації. Як приклад можна взяти налаштування `max_depth` або `min_samples_leaf`.

3.  `Випадковий ліс` знижує перенавчання шляхом навчання кількох дерев рішень та виведенням середніх значень їхніх прогнозів. також ця модель показала дещо кращі результати `R² 0,9888` і `MSE 0,0111`, що робить її надійною і кращою в узагальненні.

4.  `Метод опорних векторів (SVM)` працює дещо гірше, ніж `дерево рішень` чи `випадковий ліс`. Маючи результати `R² 0,974` і `MSE 0,0248`, SVM модель зафіксувала деяку нелінійність даних. Однак налаштування ядра та параметрів `C` може значно вплинути на продуктивність, а також модель може бути більш чутливою до масштабування порівняно з іншими.

`XGBoost Regressor` як модель посилення градієнта змогла перевершити всі вище перечислені моделі, досягнувши найвищого показника `R² (0,9888)` і найнижчого `MSE (0,0110)`. XGBoost широкий у використанні завдяки своїй продуктивності через ітераційне прискорення, плюс він може добре обробляти як лінійні, так і нелінійні шаблони, створюючи моделі послідовно для виправлення помилок попередніх.
